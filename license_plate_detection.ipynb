```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt") 
```

```python

model.train(
    data="dataset/dataset.yaml",  
    epochs=5,            
    imgsz=640,            
    batch=8,             
    device=0              
)

```

```python
model.val(data="dataset/dataset.yaml")
```

```python

img = 'D:/anpr2/image3.jpg'
results = model.predict(img)
first_result = results[0]
first_result.show()  
```

```python
import os
save_path = "D:/anpr2/runs/detect/"
os.makedirs(save_path, exist_ok=True)  
model.save(f"{save_path}train_final.pt")  
```
